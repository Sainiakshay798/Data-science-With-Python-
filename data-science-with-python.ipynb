## IMPORTING REQUIRED LIBRARIES & DATA:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

## IMPORTING DATA FROM CSV FILE:

```python
DataFrame = pd.read_csv("/kaggle/input/netflix-shows/netflix_titles.csv")
```

# As per obversation there 29% directors are missing or null. Country and cast also have 9% null values. 

```python
DataFrame.drop('director', axis=1, inplace=True)
```

```python
DataFrame =DataFrame[DataFrame['date_added'].notna()]
```

# Displaying value counts of ratings:

```python
DataFrame.rating.value_counts()
```

# Filling null values of rating with most occuring rating

```python
DataFrame.rating.fillna('TV-MA', inplace=True)
```

# Exploratory Data Analysis:
# Year wise number of released movies/shows

```python
plt.figure(figsize=(13,10))
g = DataFrame[DataFrame['release_year'] > 2010].release_year.value_counts().plot.bar(figsize=(10, 7))
g.set_xlabel("Release Year")
g.set_ylabel("No of Released Movies")
g.set_title("Year Wise Releases")
```

# Country Analysis:

```python
DataFrame.country.value_counts()
```

# Exploring Movies Vs Shows:

```python
g = DataFrame.type.value_counts().plot.pie(figsize=(10, 7),autopct= "%1.1f%%")
g.set_ylabel("Types")
g.set_title("Movies vs TV Shows")
```

# From the above pie chart we can easily conclude that number of movies released on Netflix is much higher than that of the Tv shows
# Content distribution based on rating:

```python
plt.figure(figsize=(18,7))
g = sns.countplot(x='rating', data=DataFrame, palette='Blues_r', order=DataFrame['rating'].value_counts().index[0:15])
g.set_xlabel("Ratings")
g.set_ylabel("No of Movies")
g.set_title("Ratings Wise Counts")
```

# Showing content in top 5 country wise ratings:

```python
plt.figure(figsize=(20,5))
top_countries = DataFrame[DataFrame['country'].isin(['United States', 'India', 'United Kingdom', 'Japan', 'South Korea'])]
g = sns.histplot(x='rating', y='country', data=top_countries)
g.set_xlabel("Ratings")
g.set_ylabel("Countries")
g.set_title("Top 5 country wise ratings")
```

# Top 10 oldest movies:

```python
DataFrame[DataFrame['type'] == 'Movie'].sort_values('release_year')[0:10]
```

# Top 10 oldest Tv shows:

```python
DataFrame[DataFrame['type'] == 'TV Show'].sort_values('release_year')[0:10]
```

# Extract month from date_added column

```python
DataFrame['month'] = pd.to_datetime(DataFrame['date_added']).dt.month
```

# Showing content distribution based on rating:

```python
plt.figure(figsize=(18,7))

g = sns.countplot(x='month', data=DataFrame)
g.set_xticklabels(['January','February','March','April',
                   'May','June','July','August','September','October','November','December'])
g.set_xlabel("Months")
g.set_ylabel("No of Movies")
g.set_title("Month wise releases")
```

# As we compare to other months, more movies & tv shows are released in July & December month.
